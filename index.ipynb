# How Is Attrition Affected by Gender?

```python
# Filter the dataset for rows where Attrition is "Yes"
attrition_yes = data[data['Attrition'] == 'Yes']
```

```python
# Create a histogram
fig = px.histogram(attrition_yes, x='Gender', color='Gender',
                   title='Distribution of Attrition (Yes) by Gender',
                   labels={'Gender': 'Gender', 'count': 'Count'},
                   category_orders={'Gender': attrition_yes['Gender'].unique()})

# Show the plot
fig.show()
```

# How Is Attrition Affected by business travels?

```python
plt.figure(figsize=(15,7))
sns.countplot(x='BusinessTravel', hue='Attrition', data=data, palette='seismic')
plt.title("Attrition with BusinessTravel")
plt.show()
```

# what is effect of the distance from home on attrition ?

```python
plt.figure(figsize=(15,7))
sns.countplot(x='DistanceFromHome', hue='Attrition', data=data, palette='seismic')
plt.title("Attrition with BusinessTravel")
plt.show()
```

# Attrition vs Department

```python
plt.figure(figsize=(15,7))
plt.title("Attrition with Department")
sns.countplot(x='Department', hue='Attrition', data=data);
```

# effect of YearsWithCurrManager on attrition

```python
plt.figure(figsize=(15,5))

sns.kdeplot(data=data, x = "Age", hue="Attrition",palette='Set1')
```

```python
attrition_monthincome = data[['Attrition', 'MonthlyIncome']]
```

```python
fig = px.box(attrition_monthincome, x='Attrition', y='MonthlyIncome', color='Attrition', 
             points='all',color_discrete_map={'No': 'blue', 'Yes': 'red'})

fig.update_layout(
    title='Attrition by Monthly Income',
    xaxis=dict(title='Attrition'),
    yaxis=dict(title='Monthly Income'),
    template='seaborn'
    
)

fig.show()
```
